## What is it?
A pie chart is a circular statistical graphic divided into slices to illustrate numerical proportions.

---


## Why do we use it?
 - To show the relative proportions or percentages of different categories within a whole.
 - To make comparisons between parts of a dataset.

---


## Purpose
 - To provide a visual representation of the proportion of each category as part of a whole.
 - To easily see which categories are the largest or smallest.

---


## What it tells us
 - The percentage or proportion of each category in relation to the total dataset.
 - The relative size of each category compared to others.

---

## When to Use
 - **Showing Proportions**: When you want to show how different parts make up a whole.
- Scenario: Displaying the market share of different smartphone brands.
 - **Comparing Categories**: When you want to compare the size of different categories.
- Scenario: Comparing the budget allocation for different departments in a company.
 - **Visual Simplicity**: When you need a simple and clear visual representation of proportions.
- Scenario: Showing the distribution of votes among different candidates in an election.

```python
import matplotlib.pyplot as plt
```

## Type of Data
1. Categorical Data: Data divided into distinct categories.
2. Numerical Data: The value or count associated with each category.
### Examples:
 - Market share of different companies.
 - Budget distribution across different departments.
 - Survey results showing the percentage of respondents choosing each option.

```python
x1=[10,20,30,40,50]
l1=['c','c++','java','python','R']
plt.pie(x1)
plt.show()
```

```python
x = [10,20,30,40,15]
l = ['C','C++',"python",'java','R']
```

## plt.pie(data)

```python
plt.pie(x)
plt.show()
```

## labels :
### It is used to specify the names or categories for each slice of the pie chart

```python
plt.pie(x1,labels=l1)
plt.show()
```

```python
plt.pie(x,labels = l)
plt.show()
```

## explode: Separates a specific slice from the pie chart

```python
ex1=[0.2,0.0,0.0,0.0,0.0]
plt.pie(x1,labels=l1,explode=ex1)
plt.show()
```

```python
ex = [0.2,0.0,0.0,0.0,0.0]
plt.pie(x,labels = l,explode = ex)
plt.show()
```

## color : used to specify the colors for each slice of the pie chart

```python
c=['r','b','y','g','m']

plt.pie(x1,colors=c,labels=l1)
plt.show()
```

```python
c = ['r','g','y','b','m']
plt.pie(x,labels = l,colors = c)
plt.show()
```

## autopct: Formats the numeric value on the pie slices.
## shadow: Adds a shadow to the pie chart.

```python
plt.pie(x1,labels=l1,explode=ex1,autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
plt.pie(x,labels = l,explode = ex,autopct = "%0.1f%%",shadow = True)
plt.show()
```

## radius : the radius determining its overall size of the pie chart.

```python
plt.pie(x1,labels=l1,explode=ex1,radius=1.8)
plt.show()
```

```python
plt.pie(x,labels = l,explode = ex,radius = 1.5)
plt.show()
```

## labeldistance: The radial distance at which the pie labels are drawn.

 - If it is smaller than 0 then labels goes inside the pie chart.
 - And if it is greater than 1 then labels goes further away from the pie chart.

```python
plt.pie(x1,labels=l1,labeldistance=1.2)
plt.show()
```

```python
# plt.pie(x,labels = l,labeldistance = 0.5)
# plt.show()

plt.pie(x,labels = l,labeldistance = 1.4)
plt.show()
```

## startangle: Rotates the start of the pie chart by the given angle.

```python
plt.pie(x1,labels=l1,explode=ex1,autopct="%1.1f%%",startangle=180)
plt.show()
```

```python
plt.pie(x,labels = l,explode = ex , autopct = "%0.1f%%", startangle = 180)
plt.show()
```

## rotatelabels: Rotates the labels to follow the pie slices.

```python
plt.pie(x1,labels=l1,explode=ex1,autopct='%0.1f%%',rotatelabels=True)
plt.show()
```

```python
plt.pie(x,labels = l,explode = ex , autopct = "%0.1f%%", rotatelabels = True)
plt.show()
```

## textprops: Dictionary which contains different properties of text like fontsize.

```python
plt.pie(x,labels=l1,explode=ex1,autopct='%1.1f%%',textprops={'fontsize':15})
plt.show()
```

```python
plt.pie(x,labels = l,explode = ex , autopct = "%0.1f%%", textprops = {"fontsize" : 14})
plt.show()
```

## counterclock: If False, the slices are plotted counter anticlockwise.

```python
plt.pie(x1,labels=l1,explode=ex1,autopct="%1.1f%%",counterclock=False)
plt.show()
```

```python
plt.pie(x,labels = l,explode = ex , autopct = "%0.1f%%", counterclock = False)
plt.show()
```

## wedgeprops: Dictionary of properties for the pie slices contains edgecolor , edgewidth.

```python
plt.pie(x1,labels=l1,explode=ex1,autopct="%1.1f%%",wedgeprops={'edgecolor':'black','linewidth':3})
plt.show()
```

```python
plt.pie(x,labels = l,explode = ex , autopct = "%0.1f%%", wedgeprops = {'edgecolor':'m','linewidth':4})
plt.show()
```

## legend: Adds a legend to the chart, specifying its location and properties.

```python
plt.pie(x,labels=l1,explode=ex1,autopct='%1.1f%%')
plt.title("Pie Chart")
plt.legend(loc=2)
plt.show()
```

```python
plt.pie(x,labels = l,explode = ex , autopct = "%0.1f%%")
plt.title("Pie Chart")
plt.legend(loc=2)
plt.show()
```

## Combined all the customizations (parameters) of pie chart.

```python
import matplotlib.pyplot as plt

x=[10,20,50,35,77]
l=['C','C++','Python','Java','R']

c=['r','b','y','g','gold']
exp=[0.0,0.3,0,0.1,0.0]

plt.pie(x,labels=l,colors=c,explode=exp,autopct="%1.1f%%",shadow=True,startangle=148,labeldistance=1,rotatelabels=True,counterclock=False
        ,wedgeprops={'edgecolor':'silver'},textprops={"fontsize":12,'color':'black'})

plt.title("Pie plot Draw")
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
x = [10, 20, 30, 40, 15]
labels = ['C', 'C++', 'Python', 'Java', 'R']

# Define colors for each section
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgrey']

# Define the explode parameters to "explode" the 3rd slice (i.e. 'Python')
explode = (0, 0, 0.1, 0, 0)  # only "explode" the 3rd slice (Python)

# Create a pie chart
plt.pie(x, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True,radius = 1, startangle=140,labeldistance=1,
        rotatelabels=True,counterclock=False,wedgeprops={'edgecolor': 'black'},textprops={'fontsize': 12, 'color': 'blue'})

# Add a title
plt.title("Programming languages popularities")

# Add a legend
plt.legend(loc=4)

# Display the chart
plt.show()
```
