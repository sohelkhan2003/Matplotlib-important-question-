# What are Subplots in Matplotlib?

---



---



## What:
- Subplots in Matplotlib allow you to create multiple plots within a single figure.
- Each plot is called a subplot and is positioned within a grid that you define.

## Why to Use:
- Subplots are useful when you want to compare different datasets side-by-side.
- Create a dashboard of related plots.
- Organize multiple visualizations in a structured manner within one figure.

## Purpose:
- Facilitate the visualization of multiple plots in a coherent layout.
- Make it easier to compare and analyze different aspects of data.

## When to Use:
- When you need to visualize multiple plots that are related to each other.
- Comparing different variables or datasets.
- Displaying different views of the same dataset (e.g., time series and histogram).
- Creating complex dashboards with various plots.


```python
import matplotlib.pyplot as plt
```

### **plt.subplots()** is a function that creates a figure and a grid of subplots with a single call. It returns a tuple containing the figure and an array of Axes objects (the subplots).

```python
plt.subplots()
```

```python
x1=[1,2,3,4,5]
y1=[2,6,4,8,3]
fig,ax=plt.subplots()
ax.plot(x1,y1)
```

```python
x = [1,2,3,4,5]
y = [2,6,4,8,3]
```

```python
fig , ax = plt.subplots()
ax.plot(x,y)
```

```python
ax
```

### Figsize , title , xlabel , ylabel Without using subplots.

```python
plt.figure(figsize=(9,5))

plt.plot(x1,y1)
plt.title("simple line plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.figure(figsize = (10,6))

plt.plot(x,y)
plt.title("Simple Line PLot")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

Figsize , title , xlabel , ylabel With using subplots.



```python
fig,ax=plt.subplots(figsize=(9,5))
ax.plot(x1,y1)

ax.set_title("Subplot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
fig.show()
```

```python
fig , ax = plt.subplots(figsize = (10,6))
ax.plot(x,y)

ax.set_title("Subplots")
ax.set_xlabel("X")
ax.set_ylabel("Y")

fig.show()
```

### Subplot graphs customizations same as simple plots.

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.plot(x1,y1,color='r',marker='d',markersize=15,linewidth=3,linestyle='-.')
ax.set_title("Subplot")
ax.set_xlabel('X')
ax.set_ylabel('Y')
fig.show()
```

```python
fig , ax = plt.subplots(figsize = (10,6))
ax.plot(x,y,color='r',marker='d',markersize = 15,linewidth = 3,linestyle = '-.')

ax.set_title("Subplots")
ax.set_xlabel("X")
ax.set_ylabel("Y")

fig.show()
```

### Two Rows and one column.

```python
x1 = [1,2,3,4,5]
y1 = [2,6,4,8,3]
z1 = [5,3,5,7,4]
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1)

ax[0].plot(x1,y1)
ax[1].plot(x1,z1)

fig.show()
```

```python
fig , ax = plt.subplots(nrows = 2,ncols = 1)

ax[0].plot(x,y)
ax[1].plot(x,z)

fig.show()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1)

ax[0].plot(x1,y1)
ax[0].set_title("X v/s Y Graph")
ax[0].set_xlabel("X")
ax[0].set_ylabel("Y")

ax[1].plot(x1,z1)
ax[1].set_title("X v/s Y Graph")
ax[1].set_xlabel("X")
ax[1].set_ylabel("Y")
fig.show()
```

```python
fig , ax = plt.subplots(nrows = 2,ncols = 1)

ax[0].plot(x,y)
ax[0].set_title("X vs Y graph")
ax[0].set_xlabel("X")
ax[0].set_ylabel("Y")

ax[1].plot(x,z)
ax[1].set_title("X vs Z graph")
ax[1].set_xlabel("X")
ax[1].set_ylabel("Z")

fig.show()
```

### Sharex : Share the x-axis between subplots.

```python
fig,ax =plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].plot(x1,y1,color='r',marker='o')
ax[0].set_title('x v/s y graph')
ax[0].set_xlabel("X")
ax[0].set_ylabel("Y")

ax[1].plot(x1,z1,color='b',marker='d')
ax[1].set_title('x v/s Z graph')
ax[1].set_xlabel("X")
ax[1].set_ylabel("Y")

fig.show()
```

```python
fig , ax = plt.subplots(nrows = 2,ncols = 1,sharex = True,figsize = (10,6))

ax[0].plot(x,y,color='r',marker = 'o')
ax[0].set_title("X vs Y graph")
ax[0].set_ylabel("Y")

ax[1].plot(x,z,color = 'b',marker = 'd')
ax[1].set_title("X vs Z graph")
ax[1].set_xlabel("X")
ax[1].set_ylabel("Z")

fig.show()
```

### 2 rows and 2 columns.

```python
fig,ax =plt.subplots(nrows=2,ncols=2,figsize=(10,6))

ax[0,0].plot(x1,y1)
ax[1,0].plot(x1,z1)
ax[0,1].bar(x1,y1)
ax[1,1].bar(x1,z1)

fig.show()
```

```python
fig , ax = plt.subplots(nrows = 2,ncols = 2,figsize = (10,6))

ax[0,0].plot(x,y)
ax[1,0].plot(x,z)
ax[0,1].bar(x,y)
ax[1,1].bar(x,z)
```

### 2 rows and 1 column.

```python

fig = plt.figure(figsize=(6,4))

# First subplot
ax1 = fig.add_subplot(2,1,1)
ax1.plot(x1, y1)
ax1.set_title('First plot')

# Second subplot
ax2 = fig.add_subplot(2,1,2)
ax2.plot(x1, z1)
ax2.set_title("Second plot")

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8, 4))

# Create the first subplot in a grid of 2 rows, 1 column, and in position 1
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(x, y)
ax1.set_title('First Plot')

# Create the second subplot in a grid of 2 rows, 1 column, and in position 2
ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(x, z)
ax2.set_title('Second Plot')

plt.tight_layout()  # Adjust subplots to fit in figure area.
plt.show()

```

### 2 rows and 2 columns (4 plots).

```python
fig=plt.figure(figsize=(10,8))

ax1=fig.add_subplot(2,2,1)
ax1.plot(x1,y1)
ax1.set_title("First plot")

ax2=fig.add_subplot(2,2,3)
ax2.plot(x1,z1)
ax2.set_title("Sec  plot")

ax3=fig.add_subplot(2,2,4)
ax3.bar(x1,y1)
ax3.set_title("Third plot")

ax4=fig.add_subplot(2,2,2)
ax4.bar(x1,y1)
ax4.set_title("Fourth plot")

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))

# Create the first subplot in a grid of 2 rows, 2 columns, and in position 1
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x, y)
ax1.set_title('First Plot')

# Create the second subplot in a grid of 2 rows, 2 columns, and in position 2
ax2 = fig.add_subplot(2, 2, 3)
ax2.plot(x, z)
ax2.set_title('Second Plot')

# Create the third subplot in a grid of 2 rows, 2 columns, and in position 3
ax3 = fig.add_subplot(2, 2, 2)
ax3.bar(x, y)
ax3.set_title('Third Plot')

# Create the fourth subplot in a grid of 2 rows, 2 columns, and in position 4
ax4 = fig.add_subplot(2, 2, 4)
ax4.bar(x, z)
ax4.set_title('Fourth Plot')

plt.tight_layout()  # Adjust subplots to fit in figure area.
plt.show()
```
