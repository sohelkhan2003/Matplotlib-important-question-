Q1. Mean Sepal Length Across Species
The botanists want to compare the average sepal length of the three iris
species.
• Compute the mean sepal length for each species.
• Create a line plot to show the trend of the average sepal length across
species.
• Add markers and labels to highlight the differences.


```python
mean_df=df.groupby('Species')['SepalLengthCm'].mean()
mean_df

```

```python
plt.plot(mean_df.index,mean_df.values,color='r',linestyle='dashdot',linewidth=2,label='xyz' )
plt.title(' mean sepal length for each species.')
plt.xlabel("Species")
plt.ylabel("SepalLengthCm")
plt.grid(linestyle='--',color='black',alpha=0.6)
plt.show()
```

Q2. Sepal Length Variation Across Samples
Plot the sepal length of all samples in the dataset in the order of their
appearance.
• Add a horizontal grid to emphasize the range of variation.
• Highlight samples where sepal length exceeds 7 cm.


```python
plt.plot(df.index,df['SepalLengthCm'],color='r',label='xyz')
plt.title("Sepal Length Variation Across Samples")
plt.xlabel("Index")
plt.ylabel("SepalLengthCm'")
plt.grid(linestyle='--',color='black',alpha=0.5)

high_inx=df[df['SepalLengthCm']>7].index
high_value=df.loc[high_inx,'SepalLengthCm']

plt.scatter(high_inx,high_value,color='b',marker='o',s=60,label='SepalLengthCm>7.index')
plt.legend()
plt.show()

```

Q3. Sample Count by Species
Show how many samples belong to each iris species in a bar chart.
• Use distinct colors for each bar.
• Highlight the species with the highest sample count.

```python
sepecies_count=df['Species'].value_counts()
sepecies_count
```

```python
max_sep=sepecies_count.idxmax()
base_color=['skyblue','orange','green']
color=['red' if S== max_sep else base_color[i]  for i, S in enumerate(sepecies_count)]


plt.bar(sepecies_count.index,sepecies_count.values,color=color,width=0.8,align='edge'
        ,edgecolor='black',linewidth=3,linestyle='dashdot',alpha=0.7,label='bar')
plt.title("Sample Count by Species")
plt.xlabel("Species")
plt.ylabel("Number of Samples")
plt.legend()
plt.show()
```

Q4. Sepal vs. Petal Length for Different Species
To study correlations, plot sepal length against petal length for all samples.
• Use different colors or markers to distinguish species.
• Add a legend, and discuss which species shows the strongest correlation.

```python
species=df['Species'].unique()
color=['red','blue','green']
markers=['o','d','s']

for sp,col,mar in zip(species,color,markers):
    spc=df[df['Species']==sp]

    plt.scatter(spc['SepalLengthCm'],spc['PetalLengthCm'],color=col,marker=mar,label=sp,edgecolors='black',alpha=0.5)


plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.title("Sepal Length vs Petal Length by Species")
plt.legend()
plt.show()
```
