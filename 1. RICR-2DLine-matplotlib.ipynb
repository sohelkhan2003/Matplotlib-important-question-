# 2D line plot in matplotlib



## --> What is it?     

###A 2D Line Plot is a graph that shows data points connected by straight lines, representing the relationship between two variables.


---


## --> Why do we use it?           
###To visualize trends or changes over time.
###To compare two datasets.


---


## --> Purpose :
###To help understand the relationship between variables and observe trends.    


---


## --> What it tells us :
###The direction and strength of a relationship between two variables.
###Patterns, trends, and anomalies in the data.


---

## --> When to Use
### 1. Trends Over Time: When you want to show how a variable changes over time.


*   Scenario: Plotting monthly sales data to observe trends throughout the year.


### 2. Comparing Multiple Datasets: When comparing trends of multiple variables over the same period.


*   Scenario: Comparing the temperature changes in two different cities over the same year.



### --> Importing the matplotlib.pyplot library with the alias plt

```python
import matplotlib.pyplot as plt
```

##Type of Data
###Numerical Data: Continuous data where the relationship between the data points matters.
###Examples:  
###Time series data (e.g., stock prices over time).
###Measurement data (e.g., temperature changes over days).

## Basic Code: plt.plot(x, y)
## X-axis and Y-axis: Represents the variables being compared.

```python
x = [1,2,3,4,5,6,7,8,9]   # x-data
y = [2,8,4,16,10,12,6,12,8]   # y-data
plt.plot(x,y)  # ploting the graph using x-data and y-data
```

## --> Labels and Title:
### Add using  
1. plt.xlabel()
2. plt.ylabel()
3. plt.title().

```python
plt.plot(x,y)
plt.title('Numbers count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
```

## --> show() function.

```python
plt.plot(x,y)
plt.title('Numbers count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()  # It is used to display the graph .It is automatically used by the colab but it should be used manually in other ide.
```

## Customization:

1. Line Style: Change using parameters like linestyle='--' (dashed line), '-' (solid line).
2. Line Width: Change using linewidth=2.
3. Color: Change using color='red'.
4. Markers: Add markers at data points using marker='o'
5. Marker size : change marker size like - 1,2,4,6,8..

# Parameters.

## --> Giving color to the line.

```python
plt.plot(x,y,color='red') # color gives the specified color to the line ex: color name or hex code of color
plt.title('Numbers count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

## --> linestyle : Giving line a style.    
    ex: 'dash' , 'dot' , 'dashdot' , etc

```python
plt.plot(x,y,color='red',linestyle='dashdot') # linestyle gives the style to the line ex: dotted,dsahed,solid,dashdot
plt.title('Numbers count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

## --> linewidth : Giving width to the line.

```python
plt.plot(x,y,color='red',linewidth = 3) # linewidth decides the width of line ex: 1 or 2 or 3 or 4 or 5 ...
plt.title('Numbers count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

## --> Marker : It is used highlight the edges of the graph.

```python
plt.plot(x,y,marker = 'o') # marker highlights the edges of the graph ex: 'o' or 'D' or '+' or '>' , etc
plt.title('Numbers count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

## --> markersize : This defines the size of the marker.

```python
plt.plot(x,y,marker = 'd',markersize=20)
plt.title('Numbers count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

## --> Combined graph of all the parameters.

```python
plt.plot(x,y,color='red',linestyle='-.',linewidth = 2,marker = 'o',markersize = 10)
plt.title('Numbers count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

## Multi line gragh.

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,8,4,16,10,12,6,12,8]
z = [3,12,7,10,17,11,7,13,5]
plt.plot(x,y,color='red',linestyle='-.',linewidth = 2,marker = 'o',markersize = 10)
plt.plot(x,z,color='black',linestyle='dashed',linewidth = 1,marker = 'd',markersize = 6)
plt.title('Numbers count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

## label : It decides the name of particular line.       
### legend() funcion

```python
plt.plot(x,y,label = 'abc')
plt.plot(x,z,label = 'xyz')
plt.title('Numbers count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## --> Combined gragh of all the parameters.



```python
plt.plot(x,y,color='red',linestyle='-.',linewidth = 2,marker = 'o',markersize = 10,label = 'abc')
plt.plot(x,z,color='black',linestyle='dashed',linewidth = 1,marker = 'd',markersize = 6,label = 'xyz')
plt.title('Numbers count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## --> grid() function : apply the grid to the gragh.

```python
plt.plot(x,y,color='red',linestyle='-.',linewidth = 2,marker = 'o',markersize = 10,label = 'abc')
plt.plot(x,z,color='black',linestyle='dashed',linewidth = 1,marker = 'd',markersize = 6,label = 'xyz')
plt.title('Numbers count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.legend()
plt.grid()
plt.show()
```
