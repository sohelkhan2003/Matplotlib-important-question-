Q3 The finance team wants to see how profits grew month by month. Create a bar
plot for month_number versus total_profit.
• Color-code the bars to indicate growth (e.g., green for increasing, red for
decreasing).
• Highlight the month with the highest profit.

```python
color=[]
for i in range(len(df)):
    if i==0:
        color.append("blue")
    elif df['total_profit'][i]>df['total_profit'][i-1]:
        color.append('green')
    else:
        color.append('red')
max_idx=df['total_profit'].idxmax()
color[max_idx]='gold'
    

plt.bar(df['month_number'],df['total_profit'],color=color,width=0.8,edgecolor='black',linewidth=3,alpha=0.8,label='profit')
plt.title("Rel b/w maonth num and totalprofit")
plt.xlabel("Total num ")
plt.ylabel("Total profit")
plt.legend()
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

Q7. Product Share in Sales for June 
The company wants to know the percentage share of each product in total sales for June. Create a pie chart for month 6, showing the contribution of each product. 
•	Add percentage labels to the pie slices. 
•	Highlight the product with the largest share using an exploded slice. 


```python
sales = [2760, 1555, 4890, 7490, 1890, 1555]
sales
label=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
max_index = sales.index(max(sales))
colors = ['r','b','y','orange','purple','g']
exp=[0.1 if i==max_index else 0 for i in range(len(sales))]

plt.pie(sales,labels=label,colors=colors,explode=exp,autopct='%1.1f%%',shadow=True,startangle=180,labeldistance=1,rotatelabels=True,counterclock=False,wedgeprops={'edgecolor':'black'})
plt.title("Product Share in Sales for June (Month 6)")
plt.legend(title="Products",bbox_to_anchor=(1.05,1),loc='upper left')
plt.show()

```

Q8. Monthly Sales Breakdown 
For a detailed view of monthly performance, create a subplot grid showing the monthly sales trends for each product individually. 
•	Use a consistent style for all subplots for easy comparison. 
•	Add a summary plot comparing total_units and total_profit to conclude the analysis. 

```python
sales=df
label=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']

fig =plt.figure(figsize=(12,8))

ax1=fig.add_subplot(3,3,1)
ax1.plot(df['month_number'],df['facecream'],color='r',marker='o',linestyle='dashdot')
ax1.set_title("Face cream sales")

ax2=fig.add_subplot(3,3,2)
ax2.plot(df['month_number'],df['facewash'],color='b',marker='d',linestyle='-')
ax2.set_title("face wash sales")

ax3=fig.add_subplot(3,3,3)
ax3.plot(df['month_number'],df['toothpaste'],color='g',marker='s',linestyle='--')
ax3.set_title("tooth paste sales")

ax4=fig.add_subplot(3,3,4)
ax4.plot(df['month_number'],df['bathingsoap'],color='y',marker='o',linestyle='dashdot')
ax4.set_title("bathingsoap	 sales")

ax5=fig.add_subplot(3,3,5)
ax5.plot(df['month_number'],df['shampoo'],color='blue',marker='s',linestyle='--')
ax5.set_title("shampoo	 sales")

ax6 = fig.add_subplot(3,3,6)
ax6.plot(df['month_number'], df['moisturizer'], color='y', marker='*', linestyle='-')
ax6.set_title("Moisturizer Sales")

ax7 = fig.add_subplot(3,3,7)
ax7.plot(df['month_number'], df['total_profit'], color='r', marker='o', label='Total Units')
ax7.plot(df['month_number'], df['total_profit'], color='b', marker='s', linestyle='--', label='Total Profit')
ax7.set_xlabel('Month')
ax7.set_ylabel('Profit / Units')
ax7.grid(True)
ax7.legend() 
```
